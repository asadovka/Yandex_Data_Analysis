### –£—Å—Ç—Ä–∞–Ω–µ–Ω–∏–µ –ø—Ä–æ–ø—É—Å–∫–æ–≤ –≤ –¥–∞–Ω–Ω—ã—Ö

#### –£–¥–∞–ª–µ–Ω–∏–µ –ø—Ä–æ–ø—É—Å–∫–æ–≤ name, year_of_release, genre

```python
games = games.dropna(subset = ['name', 'year_of_release', 'genre']).reset_index(drop = True)
games.info()
```

#### –û–±—Ä–∞–±–æ—Ç–∫–∞ –ø—Ä–æ–ø—É—Å–∫–æ–≤ —Å—Ç–æ–ª–±—Ü–æ–≤ critic_score, user_score, rating

```python
games['user_score'] = games['user_score'].replace('tbd', np.nan)
games['rating'] = games['rating'].replace(np.nan, 'undefined')

print('–°—Ç–æ–ª–±–µ—Ü user_score:', games['user_score'].unique(),'\n\n')
print('–°—Ç–æ–ª–±–µ—Ü rating:', games['rating'].unique(),'\n\n')
```

```python
data_1 = relevant_games.loc[relevant_games['platform']=='XOne', 'user_score'].dropna()
data_2 = relevant_games.loc[relevant_games['platform']=='PC', 'user_score'].dropna()
alpha = 0.05

#–æ—Ü–µ–Ω–∏–≤–∞–µ–º —Ä–∞–∑–Ω–∏—Ü—É –¥–∏—Å–ø–µ—Ä—Å–∏–π –≤—ã–±–æ—Ä–æ–∫
print('–ê–±—Å–æ–ª—é—Ç–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –¥–∏—Å–ø–µ—Ä—Å–∏–π –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏—Ö —Ä–µ–π—Ç–∏–Ω–≥–æ–≤ –ø–ª–∞—Ç—Ñ–æ—Ä–º —Ç–∞—Ä–∏—Ñ–æ–≤ XOne –∏ PC:', [np.var(x, ddof=1) for x in [data_1, data_2]])
print('\np-value –¥–ª—è –¥–∏—Å–ø–µ—Ä—Å–∏–π –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏—Ö —Ä–µ–π—Ç–∏–Ω–≥–æ–≤ –ø–ª–∞—Ç—Ñ–æ—Ä–º —Ç–∞—Ä–∏—Ñ–æ–≤ XOne –∏ PC:', levene(data_1, data_2, center = 'mean').pvalue,'\n')

if (levene(data_1, data_2).pvalue < alpha):
    print('–í—ã–≤–æ–¥: –¥–∏—Å–ø–µ—Ä—Å–∏–∏ –≤—ã–±–æ—Ä–æ–∫ —Ä–∞–∑–ª–∏—á–∞—é—Ç—Å—è –∑–Ω–∞—á–∏—Ç–µ–ª—å–Ω–æ. –î–ª—è –ø—Ä–æ–≤–µ—Ä–∫–∏ –≥–∏–ø–æ—Ç–µ–∑—ã –∏—Å–ø–æ–ª—å–∑—É–µ–º –ø–∞—Ä–∞–º–µ—Ç—Ä equal_var = False')
else:
    print('–í—ã–≤–æ–¥: –¥–∏—Å–ø–µ—Ä—Å–∏–∏ –≤—ã–±–æ—Ä–æ–∫ –Ω–µ —Ä–∞–∑–ª–∏—á–∞—é—Ç—Å—è –∑–Ω–∞—á–∏—Ç–µ–ª—å–Ω–æ. –î–ª—è –ø—Ä–æ–≤–µ—Ä–∫–∏ –≥–∏–ø–æ—Ç–µ–∑—ã –∏—Å–ø–æ–ª—å–∑—É–µ–º –ø–∞—Ä–∞–º–µ—Ç—Ä equal_var = True')
```

```python
results = st.ttest_ind(
    data_1,
    data_2,
    equal_var = False)

print('p-value:', results.pvalue)

if (results.pvalue < alpha):
    print('–û—Ç–≤–µ—Ä–≥–∞–µ–º –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É')
else:
    print('–ù–µ —É–¥–∞–ª–æ—Å—å –æ—Ç–≤–µ—Ä–≥–Ω—É—Ç—å –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É')
```
